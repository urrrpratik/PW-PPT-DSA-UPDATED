```python
matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]]
target = 3
def searchMatrix(matrix,target):
    rows = len(matrix)
    cols = len(matrix[0])
    row = 0
    for row in range(rows):
        if matrix[row][0] > target:
            row = row - 1
            break
    else:
        row = rows - 1
    for i in range(cols):
        if matrix[row][i] == target:
            return(True)
            break
    return(False)
searchMatrix(matrix , target)
```

## Q.3


```python
def validMountainArray(arr):
    n = len(arr)
    
    if n < 3:
        return False
    
    i = 0
    
    while i + 1 < n and arr[i] < arr[i + 1]:
        i += 1
    
    if i == 0 or i == n - 1:
        return False
    
    while i + 1 < n and arr[i] > arr[i + 1]:
        i += 1
    
    return i == n - 1

arr = [2, 1]
print(validMountainArray(arr))
```

## Q4


```python
def findMaxLength(nums):
    max_length = 0
    count = 0
    count_dict = {0: -1}  # Initialize with count 0 at index -1

    for i in range(len(nums)):
        if nums[i] == 0:
            count -= 1
        else:
            count += 1

        if count in count_dict:
            max_length = max(max_length, i - count_dict[count])
        else:
            count_dict[count] = i

    return max_length


nums = [0, 1]
print(findMaxLength(nums))
```

## Q5


```python
def minProductSum(nums1, nums2):
    nums1.sort()
    nums2.sort()
    n = len(nums1)
    min_product_sum = 0

    for i in range(n):
        min_product_sum += nums1[i] * nums2[n - i - 1]

    return min_product_sum


nums1 = [5, 3, 4, 2]
nums2 = [4, 2, 2, 5]
print(minProductSum(nums1, nums2))
```

## Q6


```python
def findOriginalArray(changed):
    from collections import Counter
    if len(changed)%2!=0:return []
    changed.sort()
    c=Counter(changed)
    ans=[]
    if c[0]%2==0:
        ans+=[0]*(c[0]//2)
        
    for i in c:
        if i==0 or c[i]==0:
            continue
        elif (i*2 not in c) or c[i]>c[i*2]:
            return []
        c[i*2]-=c[i]
        ans+=[i]*c[i]
            
    return ans


changed = [1,3,4,2,6,8]
findOriginalArray(changed)
```

## Q7


```python
def generateMatrix(n):
    result = [[0] * n for _ in range(n)]
    start_row, end_row = 0, n - 1
    start_col, end_col = 0, n - 1
    num = 1

    while start_row <= end_row and start_col <= end_col:
        for i in range(start_col, end_col + 1):
            result[start_row][i] = num
            num += 1
        start_row += 1

        for i in range(start_row, end_row + 1):
            result[i][end_col] = num
            num += 1
        end_col -= 1

        if start_row <= end_row:
            for i in range(end_col, start_col - 1, -1):
                result[end_row][i] = num
                num += 1
            end_row -= 1

        if start_col <= end_col:
            for i in range(end_row, start_row - 1, -1):
                result[i][start_col] = num
                num += 1
            start_col += 1

    return result


n = 3
print(generateMatrix(n))
```

```python
mat1 = [[1,0,0],[-1,0,3]]
mat2 = [[7,0,0],[0,0,0],[0,0,1]]
def multiply(mat1,mat2):

  m = len(mat1)
  n = len(mat2)
  l = len(mat2[0])
  ans = [[0] * l for _ in range(m)]
  nonZeroColIndicesInMat2 = [
      [j for j, a in enumerate(row) if a]
      for row in mat2
  ]

  for i in range(m):
    for j, a in enumerate(mat1[i]):
      if a == 0:
        continue
      # mat1's j-th column matches mat2's j-th row
      for colIndex in nonZeroColIndicesInMat2[j]:
        ans[i][colIndex] += a * mat2[j][colIndex]

  return ans
multiply(mat1, mat2)
```
