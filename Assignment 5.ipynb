```python
def convert_to_2d(original, m, n):
    if m * n != len(original):
        return []
    
    result = []
    for i in range(m):
        row = original[i * n : (i + 1) * n]
        result.append(row)
    
    return result


original = [1, 2, 3, 4]
m = 2
n = 2

result = convert_to_2d(original, m, n)
print(result)
```

```python
def count_complete_rows(n):
    # Start with 0 complete rows
    count = 0

    # Keep subtracting coins from n until it becomes less than the count
    while n >= count + 1:
        count += 1
        n -= count

    return count

n = 5
result = count_complete_rows(n)
print(result)
```

```python
def sorted_squares(nums):
    # Square each number in the array
    squares = [num**2 for num in nums]

    # Sort the squared numbers in non-decreasing order
    squares.sort()

    return squares

nums = [-4, -1, 0, 3, 10]
result = sorted_squares(nums)
print(result)
```

```python
def find_disjoint(nums1, nums2):
    set1 = set(nums1)
    set2 = set(nums2)

    result = [[num for num in nums1 if num not in set2],
              [num for num in nums2 if num not in set1]]

    return result


nums1 = [1, 2, 3]
nums2 = [2, 4, 6]
result = find_disjoint(nums1, nums2)
print(result)
```

```python
def distance_value(arr1, arr2, d):
    count = 0

    for num1 in arr1:
        for num2 in arr2:
            if abs(num1 - num2) <= d:
                break
        else:
            count += 1

    return count


arr1 = [4, 5, 8]
arr2 = [10, 9, 1, 8]
d = 2

result = distance_value(arr1, arr2, d)
print(result)
```

```python
def find_duplicates(nums):
    result = []

    for num in nums:
        idx = abs(num) - 1
        if nums[idx] < 0:
            result.append(abs(num))
        nums[idx] = -nums[idx]

    return result


nums = [4, 3, 2, 7, 8, 2, 3, 1]
result = find_duplicates(nums)
print(result)
```

```python
def find_minimum(nums):
    left = 0
    right = len(nums) - 1

    while left < right:
        mid = (left + right) // 2

        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid

    return nums[left]


nums = [3, 4, 5, 1, 2]
result = find_minimum(nums)
print(result)
```
